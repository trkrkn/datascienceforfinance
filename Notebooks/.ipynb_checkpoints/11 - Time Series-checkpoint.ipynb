# Time Series Analysis with Python for Finance

## 1. Overview

**Time series analysis** is crucial for financial data, as stock prices, economic indicators, and sales forecasts are often dependent on time. 

**What is Time Series Data?**

- A time series is a sequence of data points recorded at successive and equally spaced points in time.
    - Examples in Finance: Stock prices, interest rates, GDP growth, and exchange rates.

**Components of Time Series Data**
- **Trend:** Long-term increase or decrease in the data.
- **Seasonality:** Repeating patterns or cycles (e.g., sales increasing during the holiday season).
- **Noise/Residual:** Random fluctuations that are not explained by the model.

This notebook covers:

1. The **basics of time series** data and its components.
2. How to **manipulate** and **visualize** time series data with `pandas` and `matplotlib`.
3. Apply **basic time series models** such as moving averages and correlations.

### 2.3 Changes over Time

Statistical analysis methods are often based on changes over time and not the absolute values themselves. There are multiple options to calculate the changes in a time series over time, including absolute differences, percentage changes, and logarithmic (log) returns.

#### **Absolute differences**

````python
data.diff().head()
````

````python
data.diff().mean()
````

#### **Percentage changes**

````python
data.pct_change().round(3).head()
````

````python
data.pct_change().mean().plot(kind = 'bar', figsize = (10,6));
````

#### **Log Returns**

````python
rets = np.log(data / data.shift(1))
````

##### *Lets break it down:*

The expression `rets = np.log(data / data.shift(1))` is used to calculate **logarithmic returns**.

1. **`data.shift(1)`**:
   - This operation shifts the entire dataset down by one row (or period), meaning each value is replaced by the previous day's (or period's) value.
   - For example, if `data` represents daily stock prices:
     - `data.shift(1)` represents yesterday's prices for each stock.

   Example:
   ```python
   data = [100, 102, 105, 107]
   data.shift(1) = [NaN, 100, 102, 105]
   ```

2. **`data / data.shift(1)`**:
   - This divides the current value by the previous value. For time series data like stock prices, this calculates the **simple return** or **relative price change** for each period.
   - Simple return formula:
     $$
     R_t = \frac{P_t}{P_{t-1}}
     $$

   Example:
   ```python
   data / data.shift(1) = [NaN, 102/100, 105/102, 107/105] = [NaN, 1.02, 1.02941, 1.01905]
   ```

3. **`np.log()`**:
   - This takes the natural logarithm (`log`) of the result from the previous step, converting the relative price change into logarithmic returns.
   - The formula for logarithmic return:
     $$
     \text{Log Return} = \log\left(\frac{P_t}{P_{t-1}}\right)
     $$

   Example:
   ```python
   np.log([NaN, 1.02, 1.02941, 1.01905]) = [NaN, 0.01980, 0.02902, 0.01886]
   ```

4. **What it does overall:**
   - `rets = np.log(data / data.shift(1))` calculates the **logarithmic returns** for each period in the time series. Log returns are often used in financial analysis because they simplify the compounding of returns over multiple periods.

   `data` therefore contains daily stock prices, `rets` will give you the daily logarithmic returns for the stock.

````python
rets.head().round(2)
````

````python
rets.cumsum().apply(np.exp).plot(figsize = (10,6));
# np.exp since rets is log return
````

### 2.4 Resampling

Resampling of financial time series data refers to the process of converting the frequency of data points in a time series. This is commonly done in finance when you want to change the time period of financial data, for example, from daily prices to weekly or monthly prices, or from minute-by-minute trading data to hourly or daily data.

Resampling is useful for:

- Summarizing or aggregating high-frequency data (e.g., converting daily data to monthly data).
- Downsampling: Reducing the frequency of data (e.g., from daily to monthly).
- Upsampling: Increasing the frequency of data by introducing new points (e.g., from monthly to daily by interpolating data).

````python
data.resample('1w', label='right').last().head()
````

##### *Let's break it down* 

1. `data.resample('1w')`:

- The `resample()` method is used to change the frequency of the time series data.
    - `1w`: This specifies that the data should be resampled to a frequency of 1 week.
By default, pandas will resample the data with the weekly period ending on Sunday unless specified otherwise.

2. `label='right'`:

- `label='right'` controls how the labels for the resampled periods are displayed. '`right`' means the weekly period will be labeled with the end of the week (i.e., the right side of the interval). For example, if the week starts on Monday and ends on Sunday, the label will be the Sunday date.

3. `last()`:
- `last()`: This method returns the last observation in each resampled period (in this case, for each week). It is commonly used in financial data to get the closing price at the end of each week.

````python
data.resample('1m', label = 'right').last().head()
````

````python
rets.cumsum().apply(np.exp). resample('1m', label='right').last().plot(figsize=(10, 6));
````

## 3. Rolling Statistics

Let's focus on a signle financial time series

````python
sym = 'AAPL.O'
data = pd.DataFrame(data[sym]).dropna()
data.tail()
````

````python
window = 20
data['min'] = data[sym].rolling(window=window).min()
data['mean'] = data[sym].rolling(window=window).mean()
data['std'] = data[sym].rolling(window=window).std()
data['median'] = data[sym].rolling(window=window).median()
data['max'] = data[sym].rolling(window=window).max()
data['ewma'] = data[sym].ewm(halflife=0.5, min_periods=window).mean()
````

````python
data.dropna().head()
````

````python
ax = data[['min', 'mean', 'max']].iloc[-200:].plot(
    figsize = (10,6), style = ['g--', 'r--', 'g--'], lw = 0.8)
data[sym].iloc[-200:].plot(ax = ax, lw = 2.0) ;
````

#### Technical Analysis Example: SMAs 

A decades-old trading strategy based on technical analysis is using *two simple moving averages* (SMAs): The idea is that the trader should go long on a stock (or financial instrument in general) when the shorter-term SMA is above the longer-term SMA and should go short when the opposite holds true. 

````python
data['SMA1'] = data[sym].rolling(window=42).mean()
data['SMA2'] = data[sym].rolling(window=252).mean()
````

````python
data[[sym, 'SMA1', 'SMA2']].tail()
````

````python
data[[sym, 'SMA1', 'SMA2']].plot(figsize=(10, 6));
````

SMAs are then used to derive positions to implement a trading strategy. 

- a long position by a value of 1 
- a short position by a value of -1. 

The change in the position is triggered by a crossover of the two lines representing the SMA time series:

````python
data.dropna(inplace = True)
````

````python
data['positions'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)
````

````python
ax = data[[sym, 'SMA1', 'SMA2', 'positions']].plot(
    figsize = (10,6), secondary_y = 'positions')
ax.get_legend().set_bbox_to_anchor((0.25,0.85));
````

### 4.1 Logarithmic Returns

````python
rets = np.log(data / data.shift(1))
````

````python
rets.head()
````

````python
rets.dropna(inplace=True)
````

````python
rets.plot(subplots=True, figsize=(10, 6));
````

````python
pd.plotting.scatter_matrix(rets,
                           alpha=0.2,
                           diagonal='hist',
                           hist_kwds={'bins': 35},
                           figsize=(10, 6));
````

### 4.2 OLS Regression

````python
reg = np.polyfit(rets['.SPX'], rets['.VIX'], deg=1)
````

````python
ax = rets.plot(kind='scatter', x='.SPX', y='.VIX', figsize=(10, 6))
ax.plot(rets['.SPX'], np.polyval(reg, rets['.SPX']), 'r', lw=2);
````

### 4.3 Correlation

````python
rets.corr()
````

````python
ax = rets['.SPX'].rolling(window=252).corr(
    rets['.VIX']).plot(figsize=(10, 6))
ax.axhline(rets.corr().iloc[0, 1], c='r');
````
